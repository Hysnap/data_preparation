```python
import ast

def string_to_list(s):
    try:
        if isinstance(s, str):
            return ast.literal_eval(s)
        elif isinstance(s, list):
            return s
        else:
            return []
    except (ValueError, SyntaxError):
        return []
```

```python
df['locationsfromarticle2'] = (
    df['locationsfromarticle'].apply(string_to_list))

df['locationsfromarticle2'].head()
```
